### Macierz sasiedztwa

Wada macierzy sasiedztwa jest ze w przypadku malej liczby polaczen miedzy wierzchlokami, wtedy macierz staje sie macierza rzadka i zajmuje bardzo duzo pamieci. Zaleta jest natomiast to ze mozemy w latwy sposob miec dostep do wartosci dla poszczegolnych polaczen, poniewaz jest staly czas dostepu.


### Lista sasiedztwa

Lista sasiedztwa w przeciwienstwie do macierzy sasiedztwa jest uzywana w przypadku duzej ilosci wierzcholkow oraz malej ilosci polaczen. Wtedy nie zapelniamy pamieci nie potrzebnymi zerami. Wada jest natomiast to ze zlozonosc dodania, oraz pobrania z listy sasiedztwa jest $O(n)$.

# Zadanie 2

```python
import networkx as nx
import matplotlib.pyplot as plt
import random

# Losowy graf nieskierowany
d = {i: [random.randint(1, 10)  for _ in range(2)] for i in range(1, 11)}

G = nx.Graph(d)

fig = plt.figure(figsize=(12, 10))

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G, pos, node_color='g', node_size=1300)
nx.draw_networkx_edges(G, pos, width=3,alpha=0.5,edge_color='b')

labels = {i: str(i) for i in range(1, 11)}
nx.draw_networkx_labels(G, pos, labels=labels, font_size=16)

plt.title("Graf rzedu 10.")
plt.show()
```

# Zadanie 3

Algorytm przeszukujacy graf z zadania 2 w glab.

```python
def dfs(G, node, visited=[]):
    visited.append(node)
    # Jezeli wierzcholek nie jest lisciem
    if G.get(node):
        for n in G[node]:
            # Jezeli graf jest acykliczny to kazdy kolejny wierzcholek nie bedzie wczesniej odwiedzony
            if n not in visited:
                dfs(G, n, visited)
    return visited

dfs(d, 1)
```

### Graf spojny acykliczny

```python
g = {1: [2, 3], 3: [4]}

G = nx.Graph(g)

fig = plt.figure(figsize=(7, 5))

pos = nx.planar_layout(G)

nx.draw_networkx_nodes(G, pos, node_color='g', node_size=2000)
nx.draw_networkx_edges(G, pos, width=3,alpha=0.5,edge_color='b')

labels = {i: str(i) for i in range(1, 5)}
nx.draw_networkx_labels(G, pos, labels=labels, font_size=16)

plt.title("Graph of potential fraud.")
plt.show()
```
